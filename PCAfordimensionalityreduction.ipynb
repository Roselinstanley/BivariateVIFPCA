```python
quan=[]
dataset.drop(['sl_no'],inplace=True,axis=1)
dataset.drop(['salary'],inplace=True,axis=1)
```

```python
from Bivariateca import Bivariate
v=pd.DataFrame(quan)
u=Bivariate(dataset)
qual,quan=u.quanqual()
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scale=sc.fit_transform(data)
scaled=sc.transform(data)
scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca.fit(scaled)
x=pca.transform(scaled)
print(scaled.shape)
print(x.shape)
```
